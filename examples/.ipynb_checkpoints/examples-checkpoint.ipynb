```python
%load_ext autoreload
%autoreload 2
import numpy as np
from DE_library import simulate_ODE, simulate_phase_portrait, plotting
import matplotlib.pyplot as plt

rng = np.random.RandomState(1)
```

# Kuramoto oscillator

```python
t0, t1, dt = 0, 40, 0.05
t = np.arange(t0, t1, dt)

oscN = 6
X0 = rng.rand(oscN)
W = rng.rand(oscN)
K = rng.rand(oscN, oscN)

par = {'W': W, 'K': 1*K}

phi, d_phi = simulate_ODE('kuramoto', t, X0, par = par)

plt.plot(d_phi)
plt.show()
```

# Bogdanov-Takens system

```python
t0, t1, dt = 0, 5, 0.05
t = np.arange(t0, t1, dt)

X0_range = np.random.uniform(low=[-1,-1], high=[1,1], size=(100,2))
```

```python
pos, vel = simulate_phase_portrait('bogdanov_takens', t, X0_range, par = {'beta1': -0.15, 'beta2': -0.6})

ax = None
for p in pos:
    ax = plotting.trajectories(p, ax=ax, style='->', lw=1, arrowhead=.03, arrow_spacing=7, axis=False, alpha=None)
ax.axis('square')
ax.set_xlim([-1,1])
ax.set_ylim([-1,1])
```
